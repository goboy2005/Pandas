```python
#■ Q-77.   아래의 SQL을 판다스로 구현하시오 !

#SQL> select  ename, sal, job, hiredate
 #           from  emp; 

emp[ ['ename', 'sal', 'job', 'hiredate'] ] 
```

```python
#■ Q-78. 아래의 SQL 을 판다스로 구현하시오 !

#SQL>  select  ename, sal
#           from emp
#                where  job='SALESMAN';

emp[ ['ename', 'sal'] ]    [ emp['job']=='SALESMAN'  ]    
```

```python
#■ Q-79.   월급이 3000 이상인 사원들의 이름과 월급을 출력하시오 !

#SQL> select  ename, sal
#            from  emp
#               where  sal >= 3000;

emp[ ['ename', 'sal'] ]     [ emp['sal'] >= 3000  ]    
```

```python
#■ Q-80.  아래의 SQL을 판다스로 구현하시오 !

#SQL>  select  ename, sal
#          from  emp
#           where  sal  between  1000  and 3000; 


emp[ ['ename', 'sal'] ]    [ emp['sal'].between(1000,3000) ] 
```

```python
#■ Q-81.  월급이 1000 에서 3000 사이가 아닌 사원들의 이름과 월급을 출력
#하시오 !  ( not  이 판다스에서는 ~ 입니다.)

#SQL> select  ename, sal
#           from  emp
#               where  sal  not  between  1000  and 3000; 

emp[ ['ename', 'sal'] ]   [ ~ emp['sal'].between(1000,3000)   ]    
```

```python
# ■   오라클의 기타 비교 연산자와  파이썬 판다스와 비교 

#         오라클               vs          판다스 
#1. between ..  and                emp['sal'].between(1000,3000)
#2. in                                        emp['job'].isin(['SALESMAN', 'ANALYST']) 
#3. is  null                                emp['comm'].isnull()
#4. like                                     apply 함수  
```

```python
#■ Q-83.  이번에는 직업이 CLERK, SALESMAN 이 아닌 사원들의 이름과
#직업을 출력하시오 !

emp[ ['ename', 'job'] ]    [ ~ emp['job'].isin( ['CLERK', 'SALESMAN'] )  ] 
```

```python
#■ Q-84. 커미션이 null 인 사원들의 이름과 커미션을 출력하시오 !

#select  ename, comm
#     from  emp
#           where  comm  is  null ;

emp[ ['ename', 'comm'] ]   [ emp['comm'].isnull()  ] 
```

```python
#■ Q-85. 커미션이 null 이 아닌 사원들의 이름과 커미션을 출력하시오 !

#select  ename, comm
#from emp
#where  comm  is  not  null; 

emp[ ['ename', 'comm'] ]     [ ~ emp['comm'].isnull()  ]    
```
